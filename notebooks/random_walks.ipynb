# Random Walks

```python
import numpy as np 
import pandas as pd
import random as rand
import seaborn as sns
```

## One dimension

```python
# defining the number of steps 
n = 3000
  
#creating two array for containing x and y coordinate 
#of size equals to the number of size and filled up with 0's 
x = np.zeros(n)  
  
# filling the coordinates with random variables 
for i in range(1, n): 
    val = np.random.normal(0.0, 1.0) 
    x[i] = x[i - 1] + val
      
df = pd.DataFrame(np.random.randn(n, 4))
df=df.cumsum()
```

```python
df.head()
```

```python
sns.relplot(sort=False, kind='line', estimator=None, aspect=1., data=df)
sns.set(rc={"lines.linewidth": 1.0})
```

## Two dimensions

```python
# defining the number of steps 
n = 1000
  
#creating two array for containing x and y coordinate 
#of size equals to the number of size and filled up with 0's 
x = np.zeros(n) 
y = np.zeros(n) 
  
# filling the coordinates with random variables 
for i in range(1, n): 
    val = rand.randint(1, 4) 
    if val == 1: 
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif val == 2: 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif val == 3: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
      
df = pd.DataFrame({"x":x, "y":y})
```

```python
sns.relplot(x='x', y='y', sort=False, kind='line', estimator=None, aspect=1., data=df)
sns.set(rc={"lines.linewidth": 1.0})
```
